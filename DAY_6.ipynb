```python

import pandas as pd

data = {

'Name': ['John', 'Mary', 'John'],

'Sales': [1000,1500,1000]

}

df = pd.DataFrame(data)

df = df.drop_duplicates()

print(df)
```

```python
import pandas as pd

data = {

'emp_nm': ['John', 'Mary'],

'emp_sal': [25000,30000]

}

df = pd.DataFrame(data)

df.rename(columns={

'emp_nm':'Employee Name',

'emp_sal': 'Employee Salary'

}, inplace=True)

print(df)
```
