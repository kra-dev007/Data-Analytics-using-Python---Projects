```python
def calculate_grade(mark):
    if mark >= 90:
        return "A"
    elif mark >= 75:
        return "B"
    elif mark >= 50:
        return "C"
    else:
        return "Fail"

studentmarks = 25
grade = calculate_grade(studentmarks)
print("The student's grade is:", grade)
```
