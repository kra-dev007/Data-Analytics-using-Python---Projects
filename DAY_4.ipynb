```python
import numpy as np

a = np.array([1,2,3,4,5,6])
total = np.sum(a)
print(total)


#mean --> average

import numpy as np

a = np.array([6,2,1,6,3,9])
avg = np.mean(a)
print(avg)



#median() --> middle value when sorted

import numpy as np

a = np.array([6,2,1,6,3,9])
median = np.median(a)
print(median)


```

```python
import numpy as np
scores = np.array([78,85,92,67,88])
total = np.sum(scores)
avg = np.mean(scores)
maximumscore = np.max(scores)
minimumscore = np.min(scores)

print(total)
print(avg)
print(maximumscore)
print(minimumscore)
```
