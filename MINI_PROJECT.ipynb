```python
total_sales = df["Total Revenue"].sum()

print("Total Sales:", total_sales)
```

```python
product_sales = df.groupby("Total Profit") ["Total Revenue"].sum()

print(product_sales)
```

```python
product_sales.plot(kind='bar')

plt.title("Product Sales Analysis")

plt.xlabel("Products")

plt.ylabel("Sales")

plt.show()
```
